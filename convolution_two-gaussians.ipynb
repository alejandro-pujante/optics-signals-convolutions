```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
import pandas as pd
```

```python
conv = []
tau = []

a = 2
b = 4

def f(x):
    return np.exp(-a*x**2)

def g(x):
    return np.exp(-b*x**2)

for i in np.arange(-8,8,0.1):
    def h(x):
      return f(x)*g(-x-i)
    I , err= quad(h, -20, 20)
    conv.append(I)
    tau.append(i)
```

```python
plt.plot(tau , conv)
# plt.plot(x , np.sqrt(np.pi / (a+b))*np.exp( -a*b*x**2/(a+b) ) )
```

```python
x = np.arange(-8,8,0.1)

plt.plot(x , f(x))
plt.plot(x , g(x))
```

```python
data_convolution = pd.DataFrame()
g_data = pd.DataFrame(columns = x)

data_convolution['t'] = tau
data_convolution['x'] = x
data_convolution['conv'] = conv
data_convolution['f'] = f(x)
data_convolution['g'] = g(x)
```

```python
for i in range(len(tau)):
    g_data.loc[i,:] = g(-x+tau[i])

g_data
```

```python
data_convolution.to_csv('data_convolution.csv' , index = False)
```
