```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd

from matplotlib import animation, rc
rc('animation', html='html5')
```

```python
data = pd.read_csv('data_convolution.csv')
g_data = pd.read_csv('g_data.csv')
```

```python
x = data['t']
y = data['conv']
f = data['f']

plt.plot(x,f)
```

```python
# preparamos el gráfico

fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlim(( -8 , 8 )); plt.xlabel('$x$' , fontsize = '25')
ax.set_ylim(( 0 , 2.7)); plt.ylabel('Convolución' , fontsize = '25')
# ax.set_title('Convolución de dos gaussianas' , fontsize = '30')
# ax.legend( loc="upper left", title="ejercicio 4-31-app", prop={'size':40} ,frameon=False)

# se añade cada objeto de la simulación

# [punto] = ax.plot([],[],'.',color='black',markersize=20 , label = 'punto')

[rastro] = ax.plot([],[],'-',color='blue')

plt.close();


from collections import deque

rs = deque(maxlen=700)    # cantidad de rastro que deja la bola


def animate(k):
    global x , y

    r = ([x[k] , y[k]])    # puntos x , y

    rs.append(r.copy())

    # punto.set_data( x[k] , y[k] )

    rastro.set_data(*np.array(rs).T)



    return ()

animation.FuncAnimation(fig, animate, init_func=lambda:[],frames=4*40, interval=900/40, blit=True)
```

```python
# preparamos el gráfico

fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlim(( -8 , 8 ))
ax.set_ylim(( 0 , 2.7 ))
ax.set_xlabel('$x$ (cm)' , fontsize = '20')
ax.set_ylabel('$y$ (cm)' , fontsize = '20')
ax.set_title('Convolution of two gaussian functions ', fontsize= '20')
ax.grid(alpha = 0.5)


[func2] = ax.plot( [] , [], '-' , color='red' , linewidth = '2', label = 'g')
[func1] = ax.plot( [] , [], '-' , color='blue' , linewidth = '2' , label = 'f')

[rastro] = ax.plot([],[],'-',color='black', label = 'Convolution')

ax.legend( loc="upper left", prop={'size':20} ,frameon=False)

plt.close();


from collections import deque

rs = deque(maxlen=700)    # cantidad de rastro que deja la bola


def animate(k):
    global x
    global g

    r = ([ x , g_data.iloc[k] ])
    r2 = ([x[k] , y[k]])

    rs.append(r2.copy())

    func2.set_data( x , g_data.iloc[k] )
    func1.set_data( x , f )

    rastro.set_data(*np.array(rs).T)

    return ()

animation.FuncAnimation(fig , animate , init_func=lambda:[] , frames=4*40, interval=1100/40  , blit=True)
```
